```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from collections import defaultdict
```

```python
pd.set_option("display.max_rows", None)
```

```python
%matplotlib inline
```

# Mother Qualification and Previous Qualification vs Target

```python
mother_qual_df = pd.read_csv('reformatted_mother_previous_qual_groupby_result.csv')
```

```python
mother_qual_df.head()
```

```python
mother_qual_df = mother_qual_df.sort_values(by=['previous qualification', 'mother qualification', 'target'])
mother_qual_df
```

```python
mother_qualification_target_count = defaultdict(list)

for index, row in mother_qual_df.iterrows():
  key = f"{row['previous qualification']}-{row['target']}"
  mother_qualification_target_count[key].append(row['count'])
mother_qualification_target_count
```

```python
def create_mother_qual_graph():
  groups = ['6th/7th/8th Year', '9th/10th/11th Year', '4th/5th year', 'Degree', '12th Year']

  with sns.axes_style("white"):
    sns.set_style("ticks")
    sns.set_context("talk")

    bar_width = 0.10
    epsilon = .015
    line_width = 1
    opacity = 0.7
    bar1_positions = np.arange(5)
    bar2_positions = bar1_positions + bar_width
    bar3_positions = bar2_positions + bar_width
    bar4_positions = bar3_positions + bar_width
    bar5_positions = bar4_positions + bar_width
    bar6_positions = bar5_positions + bar_width
    bar7_positions = bar6_positions + bar_width
    plt.figure(figsize=(12, 12))
    # Basic Education 3rd Cycle (9th/10th/11th Year)
    plt.bar(bar1_positions, mother_qualification_target_count['Basic Education 3rd Cycle (9th/10th/11th Year)-Dropout'], bar_width,
                              color='#548282',
                              label='9th/10th/11th Year-Dropout')
    plt.bar(bar1_positions, mother_qualification_target_count['Basic Education 3rd Cycle (9th/10th/11th Year)-Enrolled'], bar_width-epsilon,
                              bottom=mother_qualification_target_count['Basic Education 3rd Cycle (9th/10th/11th Year)-Dropout'],
                              alpha=opacity,
                              color='white',
                              edgecolor='#548282',
                              linewidth=line_width,
                              hatch='//',
                              label='9th/10th/11th Year-Enrolled')
    plt.bar(bar1_positions, mother_qualification_target_count['Basic Education 3rd Cycle (9th/10th/11th Year)-Graduate'], bar_width-epsilon,
                               bottom=np.array(mother_qualification_target_count['Basic Education 3rd Cycle (9th/10th/11th Year)-Enrolled'])+np.array(mother_qualification_target_count['Basic Education 3rd Cycle (9th/10th/11th Year)-Dropout']),
                               alpha=opacity,
                               color='white',
                               edgecolor='#548282',
                               linewidth=line_width,
                               hatch='0',
                               label='9th/10th/11th Year-Graduate')


    # Higher Education - Degree
    plt.bar(bar2_positions, mother_qualification_target_count['Higher Education - Degree-Dropout'], bar_width,
                              color='#5E5A58',
                              label='Degree-Dropout')
    plt.bar(bar2_positions, mother_qualification_target_count['Higher Education - Degree-Enrolled'], bar_width-epsilon,
                              bottom=mother_qualification_target_count['Higher Education - Degree-Dropout'],
                              color="white",
                              hatch='//',
                              edgecolor='#5E5A58',
                              ecolor="#5E5A58",
                              linewidth=line_width,
                              label='Degree-Enrolled')
    plt.bar(bar2_positions, mother_qualification_target_count['Higher Education - Degree-Graduate'], bar_width-epsilon,
                               bottom=np.array(mother_qualification_target_count['Higher Education - Degree-Enrolled'])+np.array(mother_qualification_target_count['Higher Education - Degree-Dropout']),
                               color="white",
                               hatch='0',
                               edgecolor='#5E5A58',
                               ecolor="#5E5A58",
                               linewidth=line_width,
                               label='Degree-Graduate')

    # Higher Education - Master's
    plt.bar(bar3_positions, mother_qualification_target_count["Higher Education - Master's-Dropout"], bar_width,
                              color='#A4CACA',
                              label="Master's-Dropout")
    plt.bar(bar3_positions, mother_qualification_target_count["Higher Education - Master's-Enrolled"], bar_width-epsilon,
                              bottom=mother_qualification_target_count["Higher Education - Master's-Dropout"],
                              color="white",
                              hatch='//',
                              edgecolor='#A4CACA',
                              ecolor="#A4CACA",
                              linewidth=line_width,
                              label="Master's-Enrolled")
    plt.bar(bar3_positions, mother_qualification_target_count["Higher Education - Master's-Graduate"], bar_width-epsilon,
                               bottom=np.array(mother_qualification_target_count["Higher Education - Master's-Enrolled"])+np.array(mother_qualification_target_count["Higher Education - Master's-Dropout"]),
                               color="white",
                               hatch='0',
                               edgecolor='#A4CACA',
                               ecolor="#A4CACA",
                               linewidth=line_width,
                               label="Master's-Graduate")

    # Other
    plt.bar(bar4_positions, mother_qualification_target_count['Other-Dropout'], bar_width,
                              color='#FF8B8B',
                              label='Other-Dropout')
    plt.bar(bar4_positions, mother_qualification_target_count['Other-Enrolled'], bar_width-epsilon,
                              bottom=mother_qualification_target_count['Other-Dropout'],
                              color="white",
                              hatch='//',
                              edgecolor='#FF8B8B',
                              ecolor="#FF8B8B",
                              linewidth=line_width,
                              label='Other-Enrolled')
    plt.bar(bar4_positions, mother_qualification_target_count['Other-Graduate'], bar_width-epsilon,
                               bottom=np.array(mother_qualification_target_count['Other-Enrolled'])+np.array(mother_qualification_target_count['Other-Dropout']),
                               color="white",
                               hatch='0',
                               edgecolor='#FF8B8B',
                               ecolor="#FF8B8B",
                               linewidth=line_width,
                               label='Other-Graduate')

    # Professional higher technical course
    plt.bar(bar5_positions, mother_qualification_target_count['Professional higher technical course-Dropout'], bar_width,
                              color='#FFC000',
                              label='Professional higher technical course-Dropout')
    plt.bar(bar5_positions, mother_qualification_target_count['Professional higher technical course-Enrolled'], bar_width-epsilon,
                              bottom=mother_qualification_target_count['Professional higher technical course-Dropout'],
                              color="white",
                              hatch='//',
                              edgecolor='#FFC000',
                              ecolor="#FFC000",
                              linewidth=line_width,
                              label='Professional higher technical course-Enrolled')
    plt.bar(bar5_positions, mother_qualification_target_count['Professional higher technical course-Graduate'], bar_width-epsilon,
                               bottom=np.array(mother_qualification_target_count['Professional higher technical course-Enrolled'])+np.array(mother_qualification_target_count['Professional higher technical course-Dropout']),
                               color="white",
                               hatch='0',
                               edgecolor='#FFC000',
                               ecolor="#FFC000",
                               linewidth=line_width,
                               label='Professional higher technical course-Graduate')

    # Secondary Education - 12th Year of Schooling
    plt.bar(bar6_positions, mother_qualification_target_count['Secondary Education - 12th Year of Schooling-Dropout'], bar_width,
                              color='#004771',
                              label='12th Year of Schooling-Dropout')
    plt.bar(bar6_positions, mother_qualification_target_count['Secondary Education - 12th Year of Schooling-Enrolled'], bar_width-epsilon,
                              bottom=mother_qualification_target_count['Secondary Education - 12th Year of Schooling-Dropout'],
                              color="white",
                              hatch='//',
                              edgecolor='#004771',
                              ecolor="#004771",
                              linewidth=line_width,
                              label='12th Year of Schooling-Enrolled')
    plt.bar(bar6_positions, mother_qualification_target_count['Secondary Education - 12th Year of Schooling-Graduate'], bar_width-epsilon,
                               bottom=np.array(mother_qualification_target_count['Secondary Education - 12th Year of Schooling-Enrolled'])+np.array(mother_qualification_target_count['Secondary Education - 12th Year of Schooling-Dropout']),
                               color="white",
                               hatch='0',
                               edgecolor='#004771',
                               ecolor="#004771",
                               linewidth=line_width,
                               label='12th Year of Schooling-Graduate')

    # Technological specialization course
    plt.bar(bar7_positions, mother_qualification_target_count['Technological specialization course-Dropout'], bar_width,
                              color='#745974',
                              label='Technological specialization course-Dropout')
    plt.bar(bar7_positions, mother_qualification_target_count['Technological specialization course-Enrolled'], bar_width-epsilon,
                              bottom=mother_qualification_target_count['Technological specialization course-Dropout'],
                              color="white",
                              hatch='//',
                              edgecolor='#745974',
                              ecolor="#745974",
                              linewidth=line_width,
                              label='Technological specialization course-Enrolled')
    plt.bar(bar7_positions, mother_qualification_target_count['Technological specialization course-Graduate'], bar_width-epsilon,
                               bottom=np.array(mother_qualification_target_count['Technological specialization course-Enrolled'])+np.array(mother_qualification_target_count['Technological specialization course-Dropout']),
                               color="white",
                               hatch='0',
                               edgecolor='#745974',
                               ecolor="#745974",
                               linewidth=line_width,
                               label='Technological specialization course-Graduate')

    plt.xticks(bar7_positions, groups, rotation=45)
    plt.ylabel('Count')
    plt.legend(bbox_to_anchor=(0.5, 1.15), ncol=2)
    sns.despine()
    plt.show()
```

```python
create_mother_qual_graph()
```
